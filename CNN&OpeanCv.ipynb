```
import cv2
from google.colab.patches import cv2_imshow

image = cv2.imread("/content/transferir (1).jpg")

print(type(image))
(h, w, d) = image.shape
print(image.shape)
print("Height: ", h)
print("Width: ", w)
print("Depth: ", d)
```

```
cv2_imshow(image)
```

```
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```
cv2_imshow(rgb_image)
```

```
cropped = image[10:500, 500:2000]
cv2_imshow(cropped)
```

```
width = int(w * 20 / 100)
height = int(h * 20 / 100)
dim = (width, height)
resized = cv2.resize(image, dim, interpolation = cv2.INTER_AREA)

cv2_imshow(resized)
```

```
(h, w, d) = image.shape
center = (w // 2, h // 2)
M = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated = cv2.warpAffine(image, M, (w, h))
cv2_imshow(rotated)
```

```
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

ret, threshold_image = cv2.threshold(image, 127, 255, 0)

cv2_imshow(gray_image)
cv2_imshow(threshold_image)
```

```
from matplotlib import pyplot as plt
import numpy as np

titles = ['Original', 'Gray', 'Threshold']
images = [image, gray_image, threshold_image]

for i in range(3):
  plt.subplot(2,3,i+1),
  plt.imshow(images[i],'gray',vmin=0,vmax=255)
  plt.title(titles[i])
  plt.show()
```
